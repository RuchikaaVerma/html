Column Management


```python
df=pd.read_csv('customers-100.csv')

df["First Name"].unique()
```

```python
cols_to_drop = ['Subscription Date', 'Website']
existing_cols = [col for col in cols_to_drop if col in df.columns]
df = df.drop(columns=existing_cols)
print("Columns dropped successfully.")
```

```python
df = df.rename(columns={'Customer Id': 'ID', 'Country': 'Location'})
print(df.head())
```

```python
# Example: Reorder columns
df = df.reindex(columns=['ID', 'Location', 'Age'])

# Example: Reset row index (optional)
df = df.reset_index(drop=True)
```

Missing Value Handling


```python
print(df.isna().sum())
```

```python
df_cleaned = df.dropna()
print(df_cleaned.head())
```

```python
df_filled = df.fillna(0)  # For numeric columns

print(df_filled.head())
```

```python

df_filled = df.fillna("Unknown")  # For categorical columns
print(df_filled.head())
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df['Age'] = imputer.fit_transform(df[['Age']])
print(df.head())
```
